```python
import sqlite3
```

```python
db_file = 'property.db'
connect = sqlite3.connect(db_file)
```

```python
property_details.to_sql('Property_Details', connect, if_exists='replace', index=False)
```

```python
property_price_details.to_sql('Property_Price_Details', connect, if_exists='replace', index=False)
```

```python
# Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos"

query1 = "SELECT pd.id, start_date, end_date, created_on, lat, lon, l1, l2, l3, rooms, bedrooms, bathrooms, surface_total, surface_covered, price, currency, price_period, property_type, operation_type FROM Property_Details pd JOIN Property_Price_Details ppd on pd.id = ppd.id WHERE ppd.price > 1000000 AND pd.l1 = 'Estados Unidos';"
pd.read_sql_query(query1, connect)
```

```python
# Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 'Medium' if it's between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters

query2 = "SELECT *, CASE WHEN surface_total < 50 THEN 'Small' WHEN surface_total >= 50 AND surface_total <= 100 THEN 'Medium' WHEN surface_total > 100 THEN 'Large' ELSE 'Unknown' END AS surface_area_category FROM Property_Details;"
pd.read_sql_query(query2, connect)
```

```python
# List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and bathrooms as another property in the dataset

query3 = "SELECT DISTINCT pd1.id AS property_id FROM Property_Details pd1 JOIN Property_Details pd2 ON pd1.bedrooms = pd2.bedrooms AND pd1.bathrooms = pd2.bathrooms AND pd1.l3 = 'Belgrano' AND pd2.l3 = 'Belgrano' AND pd1.id <> pd2.id;"
pd.read_sql_query(query3, connect)
```

```python
# Calculate the average price per square meter (price / surface_total) for each property type (property_type) in the "Belgrano" neighborhood (l3)

query4 ="SELECT property_type, AVG(price / surface_total) AS avg_price_per_sq_meter FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id WHERE l3 = 'Belgrano' GROUP BY property_type;"
pd.read_sql_query(query4, connect)
```

```python
# Identify properties that have a higher price than the average price of properties with the same number of bedrooms and bathrooms.

query5 = "WITH AvgPriceByBedroomBathroom AS (SELECT pd.bedrooms, pd.bathrooms, AVG(ppd.price) AS avg_price FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id GROUP BY pd.bedrooms, pd.bathrooms) SELECT pd.id, pd.bedrooms, pd.bathrooms, ppd.price FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id JOIN AvgPriceByBedroomBathroom ap ON pd.bedrooms = ap.bedrooms AND pd.bathrooms = ap.bathrooms WHERE ppd.price > ap.avg_price;"
pd.read_sql_query(query5, connect)
```

```python
# Calculate the cumulative price for each property type, ordered by the creation date.

query6 = "SELECT pd.id, property_type, created_on, price, SUM(price) OVER (PARTITION BY property_type ORDER BY created_on) AS cumulative_price FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id ORDER BY property_type, created_on;"
pd.read_sql_query(query6, connect)
```

```python
# Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for sale (operation_type = 'Venta')

query7 = "SELECT l3 AS location, SUM(pd.surface_total) AS total_surface_area FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id WHERE operation_type = 'Venta' GROUP BY l3 ORDER BY total_surface_area DESC LIMIT 10;"
pd.read_sql_query(query7, connect)
```

```python
# Find the top 5 most expensive properties (based on price) in the "Palermo" neighborhood (l3) that were listed in August 2020

query8 = "SELECT pd.id, l3 AS neighborhood, price FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id WHERE l3 = 'Palermo' AND created_on like '2020-08%' ORDER BY price DESC LIMIT 5;"
pd.read_sql_query(query8, connect)
```

```python
# Find the top 3 properties with the highest price per square meter (price divided by surface area) within each property type

query9 = "WITH RankedProperties AS (SELECT pd.id, property_type, price, surface_total, price / surface_total AS price_per_sq_meter, ROW_NUMBER() OVER (PARTITION BY property_type ORDER BY price / surface_total DESC) AS row_num FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id) SELECT id, property_type, price, surface_total, price_per_sq_meter FROM RankedProperties WHERE row_num <= 3;"
pd.read_sql_query(query9, connect)
```

```python
# Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for properties listed for sale (operation_type = 'Venta') in the year 2020. Exclude locations with fewer than 10 properties listed for sale in 2020 from the results

query10 = "WITH LocationSummary AS ( SELECT l1, l2, l3, COUNT(*) AS property_count, AVG(price / surface_total) AS avg_price_per_sq_meter FROM Property_Details pd JOIN Property_Price_Details ppd ON pd.id = ppd.id WHERE operation_type = 'Venta' AND created_on LIKE '2020%' GROUP BY l1, l2, l3) SELECT l1, l2, l3, avg_price_per_sq_meter FROM LocationSummary WHERE property_count >= 10 ORDER BY avg_price_per_sq_meter DESC LIMIT 3;"
pd.read_sql_query(query10, connect)
```
